```python
# creating spilt csv
df1 = pd.read_csv("pollution_clean.csv")
df1.head()
```

```python
df1
```

```python
df1sub = df1.loc[:,['State', 'County', 'State Code', 'County Code']]
df1sub
```

```python
# dfloc = df1sub['State Code'].astype(str) + df1sub['County Code'].astype(str)
# dfloc
# dd = df1sub['State Code']

dfloc = df1sub.drop_duplicates()
dfloc = dfloc.astype({'State Code': str, 'County Code': str})
dfloc

dfloc = dfloc.replace({'State Code': {'0': '0', '1': '01', '2': '02', '3': '03', '4': '04', '5': '05', '6': '06', '7': '07', '8': '08', '9': '09'}})
dfloc
# for i in df1sub['State Code']:
#   if i > 100:
#     print(i)

# dffips = dfloc['State Code'] + '0' + dfloc['County Code']
# dffips[1] = '04001'
# dffips

fips_list = []
for i in dfloc['County Code']:
  if int(i) < 10:
    fips_list.append("00" + i)
  elif int(i) < 100:
    fips_list.append("0" + i)
  else:
    fips_list.append(i)
# print(fips_list)
# print(len(fips_list))
dfloc['County Code'] = fips_list

dffips = dfloc['State Code'] + dfloc['County Code']
dffips

dfloc['FIPS'] = dffips
dfloc
```

```python
i=dfloc[dfloc['State']=='Wyoming']
i
```

```python
df_final_loc = dfloc.loc[:,['FIPS', 'State', 'County', 'State Code', 'County Code']]
# df_final_loc.dtypes
df_final_loc.isna().sum()
df_final_loc.to_csv("pollution_location_table.excel", index=False)
df_final_loc
```

```python
df_final_loc.to_csv("pollution_location_table.txt", index=False)
```

```python
each_line = []
with open('pollution_location_table.txt') as f:
    lines = f.readlines()
    for line in lines:
      line_split = line.split(',')
      txt = "('" + line_split[0] + "','" + line_split[1] + "','" + line_split[2] + "','" + line_split[3] + "','" + line_split[4][:-1] + "'),"
      each_line.append(txt)
f.close()

with open('location_for_sql.txt', 'w') as f1:
    for line in each_line:
        f1.write(line)
        f1.write('\n')
f1.close()
      
```
